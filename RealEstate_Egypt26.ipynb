Property Type Classification (Feature Engineering) — transforming and organizing property categories into structured features for analysis and modeling.


```python
def categorize_property(x):
    if x in ['Apartment', 'Duplex', 'Penthouse', 'Roof']:
        return 'Residential'

    elif x in ['Villa', 'Twin House', 'Townhouse', 'iVilla']:
        return 'Villa_Compound'

    elif x in ['Chalet', 'Cabin']:
        return 'Coastal'

    elif x in ['Hotel Apartment', 'Whole Building', 'Bulk Sale Unit',
               'Bulk Rent Unit', 'Full Floor', 'Half Floor']:
        return 'Investment'

    elif x in ['Land', 'Palace', 'Bungalow']:
        return 'Land_Luxury'

    else:
        return 'Unknown'


# تطبيق
df['property_group'] = df['property_type'].fillna('Unknown').apply(categorize_property)
```

Filtering Unrealistic Data — removing inconsistent, invalid, or illogical records to improve data quality and analysis accuracy.


```python
df = df[
    (
        ((df['category'] == 'buy') & (df['price_egp'] >= 500_000)) |
        ((df['category'] == 'rent') & (df['price_egp'] >= 5_000))
    )
    &
    (df['area_value'] >= 30)
]

df = df.dropna(subset=['price_egp', 'area_value'])
```

Price and Area Categorization — grouping prices and property areas into meaningful ranges to simplify analysis and identify market patterns.


```python
# ================================
# Price Category
# ================================

# فئات البيع
buy_bins   = [0, 5_000_000, 10_000_000, 20_000_000, float('inf')]
buy_labels = ['Budget', 'Mid-Range', 'Premium', 'Luxury']

# فئات الإيجار
rent_bins   = [0, 30_000, 70_000, 150_000, float('inf')]
rent_labels = ['Budget', 'Mid-Range', 'Premium', 'Luxury']

# تطبيق
df['price_category'] = None

buy_mask  = df['category'] == 'buy'
rent_mask = df['category'] == 'rent'

df.loc[buy_mask, 'price_category'] = pd.cut(
    df.loc[buy_mask, 'price_egp'],
    bins   = buy_bins,
    labels = buy_labels
)

df.loc[rent_mask, 'price_category'] = pd.cut(
    df.loc[rent_mask, 'price_egp'],
    bins   = rent_bins,
    labels = rent_labels
)

# ================================
# Area Category
# ================================

area_bins   = [0, 100, 200, 350, float('inf')]
area_labels = ['Small', 'Medium', 'Large', 'X-Large']

df['area_category'] = pd.cut(
    df['area_value'],
    bins   = area_bins,
    labels = area_labels
)

# ================================
# النتائج
# ================================

print("=" * 40)
print("Price Category:")
print(df.groupby(['category', 'price_category'])
        .size()
        .reset_index(name='count')
        .to_string(index=False))

print("\n" + "=" * 40)
print("Area Category:")
print(df.groupby(['category', 'area_category'])
        .size()
        .reset_index(name='count')
        .to_string(index=False))
```

Cleaning City and District Columns — correcting spelling mistakes in city names, standardizing naming formats, and grouping districts with fewer than 50 listings into an “Other” category to reduce noise and improve analysis quality.


```python
df['city'] = df['city'].replace({
    'South Sainai' : 'South Sinai',
    'Al Daqahlya'  : 'Al Dakahlia',
    'Demyat'       : 'Damietta'
})

df['city'] = df['city'].fillna('Unknown')
```

```python
df['town'] = df['town'].fillna('Unknown')
```

```python
df['district'] = df['district'].str.strip().str.title()
# 'Mountain view' → 'Mountain View'
```

```python
df['district'] = df['district'].fillna('Unknown')
```

```python
# أي district عندها أقل من 50 إعلان → Other
counts = df['district'].value_counts()
rare   = counts[counts < 50].index

df['district'] = df['district'].replace(
    dict.fromkeys(rare, 'Other')
)
```

Cleaning Completion Status — transforming the `completion_status` column into two clearer and more meaningful features:

* `ready_status`: indicates whether the property is ready for delivery or still under construction (`Ready` / `Off Plan`).
* `source_type`: identifies whether the property is listed directly by the developer or by an owner/reseller (`Primary` / `Secondary`).


```python
# نشوف الـ nan بتيجي من category إيه
print(df[df['completion_status'].isna()]
      ['category']
      .value_counts())
```

```python
# الإيجار → Unknown طبيعي
# البيع   → 39 صف بس، نحذفهم بدل ما نغلط في التصنيف

# حذف الـ 39 صف في البيع اللي عندهم nan
df = df[~((df['category'] == 'buy') &
          (df['completion_status'].isna()))]

# تبسيط القيم لعمودين
df['ready_status'] = df['completion_status'].map({
    'completed'         : 'Ready',
    'completed_primary' : 'Ready',
    'off_plan'          : 'Off Plan',
    'off_plan_primary'  : 'Off Plan'
})
df['ready_status'] = df['ready_status'].fillna('Unknown')

df['source_type'] = df['completion_status'].map({
    'completed'         : 'Secondary',
    'completed_primary' : 'Primary',
    'off_plan'          : 'Secondary',
    'off_plan_primary'  : 'Primary'
})
df['source_type'] = df['source_type'].fillna('Unknown')
```

Cleaning Furnishing and Payment Method — standardizing furnishing and payment categories, while assigning missing payment values in rental properties as “Cash” since rent is typically paid on a monthly basis.


```python
print(df[df['furnished'].isna()]
      ['category'].value_counts())
```

```python
print(df[df['payment_method'].isna()]
      ['category'].value_counts())
```

```python
df['furnished'] = df['furnished'].map({
    'YES'    : 'Furnished',
    'NO'     : 'Unfurnished',
    'PARTLY' : 'Partly Furnished'
})

# buy  nan → 'Not Specified'
# rent nan → 'Not Specified'
# الاتنين نفس المعاملة
df['furnished'] = df['furnished'].fillna('Not Specified')
```

```python
df['payment_method'] = df['payment_method'].map({
    'cash'                : 'Cash',
    'installments'        : 'Installments',
    'cash | installments' : 'Both'
})

# rent nan → 'Cash' (منطقي، الإيجار بيدفع شهري)
df.loc[(df['category'] == 'rent') &
       (df['payment_method'].isna()),
       'payment_method'] = 'Cash'

# buy nan → 'Not Specified' (50 صف بس)
df.loc[(df['category'] == 'buy') &
       (df['payment_method'].isna()),
       'payment_method'] = 'Not Specified'
```

Amenities Extraction
Transforming the amenities column — which contains all property features as text — into separate binary columns (0 or 1) indicating whether a feature exists or not, such as has_balcony, has_private_pool, etc.
It also creates an amenities_count column to represent the total number of available amenities for each property.

```python
# استخراج أهم المميزات كـ binary columns
amenities_features = [
    'Balcony',
    'Private Pool',
    'Shared Pool',
    'Private Garden',
    'Covered Parking',
    'Security',
    'Shared Gym',
    'Shared Spa',
    'Maids Room',
    'Central A/C',
    'Built in Wardrobes',
    'Kitchen Appliances',
    'View of Water',
    'View of Landmark',
    "Children's Pool",
    'Lobby in Building'
]

for feature in amenities_features:
    col_name = feature.lower()\
                      .replace(' ', '_')\
                      .replace("'", '')
    df[f'has_{col_name}'] = df['amenities']\
        .str.contains(feature, na=False)\
        .astype(int)

# عمود لعدد المميزات الكلي
df['amenities_count'] = df['amenities']\
    .str.split('|')\
    .str.len()\
    .fillna(0)\
    .astype(int)

# حذف العمود الأصلي
df = df.drop(columns=['amenities'])
```

Cleaning Bedrooms and Bathrooms — converting values like `"studio"` to `0` and `"7+"` to `7`, while filling missing values using the median based on the property type, which provides a smarter and more accurate approach than using a global median.


```python
df['bedrooms'] = (
    df['bedrooms']
    .replace({'studio': 0, '7+': 7})
    .pipe(pd.to_numeric, errors='coerce')
    .astype('Int64')  # نوع integer يدعم NaN
)
```

```python
df["bedrooms"] = df.groupby("property_type")["bedrooms"]\
                   .transform(lambda x: x.fillna(x.median()))
```

```python
df["bedrooms"] = df["bedrooms"].round().astype("Int64")
```

```python
df['bathrooms'] = (
    df['bathrooms']
    .replace({'none': 0, '7+': 7})
    .pipe(pd.to_numeric, errors='coerce')
    .astype('Int64')  # نوع integer يدعم NaN
)
```

```python
df["bathrooms"] = df.groupby("property_type")["bathrooms"]\
    .transform(lambda x: x.fillna(round(x.median())))
```
